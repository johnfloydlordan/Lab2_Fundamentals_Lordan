```python
# TASK 1
target_name = "LORDAN"
SEED_NUM = 2
redacted_name = " "
```

```python
for char in target_name:
    if char in "AEIOU":
        redacted_name += str(SEED_NUM)
    else:
        redacted_name += char
```

```python
print("Original Surname:", target_name)
print("SEED_NUM Used: ", SEED_NUM)
print("Redacted Output: ", redacted_name)
```

```python
# TASK 2
engineer_name = "LORDAN"
first_letter = engineer_name[0]

ascii_value = ord(first_letter)
loops = len(target_name)
```

```python
verification_hash = (ascii_value * SEED_NUM) - loops
```

```python
print("Engineer Name: ", engineer_name)
print("First Letter: ", first_letter)
print("ASCII Value: ", ascii_value)
print("Total Loops: ", loops)
print("SEED_NUM Used: ", SEED_NUM)
print("Verification Hash: ", verification_hash)
```

```python
# TASK 3
correct_password = str(verification_hash)
attempts = 0
```

```python
while True:
    user_input = input("Enter password: ")
    attempts += 1
    
    if user_input == correct_password:
        print("Access Granted.")
        break
    else:
        print("Incorrect password. Try again.")
        
```

```python
print("Correct Password Set: ", correct_password)
print("Attempts Before Success: ", attempts)
```
